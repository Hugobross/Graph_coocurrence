```python
def get_lemma(word):
    lemma = wordnet.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer

def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
def cleaning(text):
    
    STOP_WORDS = stopwords.words()
    
    text = str(text)
    text = "".join(text)
    text = text.lower()
    text = ''.join( x for x in text if x not in characters)
    text = re.sub(r"""[!?'".<>(){}@%&*#/[/]""", " ", text)
    emoji = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    text = emoji.sub(r'', text) # no emoji
    
    # removing the stop - words
    text_tokens = word_tokenize(text)
    tokens_without_sw = [
     word
     for word in text_tokens
     if not word in STOP_WORDS
    ]
    tokens_without_sw = [get_lemma(token) for token in tokens_without_sw]
    tokens_without_sw = [get_lemma2(token) for token in tokens_without_sw]

    filtered_sentence = (" ").join(tokens_without_sw)
    text = filtered_sentence
    return text

text = cleaning(text)
text
text = text.split(", ")
text
```

```python
from collections import defaultdict

def co_occurrence(sentences, window_size):
    d = defaultdict(int)
    vocab = set()
    for text in sentences:
        # preprocessing (use tokenizer instead)
        text = text.lower().split()
        # iterate over sentences
        for i in range(len(text)):
            token = text[i]
            vocab.add(token)  # add to vocab
            next_token = text[i+1 : i+1+window_size]
            for t in next_token:
                key = tuple( sorted([t, token]) )
                d[key] += 1
    
    # formulate the dictionary into dataframe
    vocab = sorted(vocab) # sort vocab
    df = pd.DataFrame(data=np.zeros((len(vocab), len(vocab)), dtype=np.int16),
                      index=vocab,
                      columns=vocab)
    for key, value in d.items():
        df.at[key[0], key[1]] = value
        df.at[key[1], key[0]] = value
    return df
```

```python
mots = " ".join(text).split()
w = len(mots)
    
    
df = co_occurrence(text, w)
df
```
